## find_element()
- 셀레늄 드라이버 객체가 가지는 find_element() 메소드는 요소를 찾기 위해 사용된다.

- 반환값 : 일치하는 요소
- 하나의 요소만 : find_element()
- 여러 요소 : find_elements()

- 사용 : 드라이버객체.find_element(By.선택자종류, '선택자')

#### By와 선택자

- from selenium.webdriver.common.by import By

- find_element() 메소드가 받는 두가지 인자에서 첫번째 자리에 들어가는 인자는 어떤 선택자를 사용할 것인지 명시하는 것이다.
- 이때 selenium 모듈에 포함된 By를 사용해야 한다.

- By가 가지는 선택자 종류 : By.XPATH, By.ID, By.NAME, By.TAG_NAME, By.CLASS_NAME, By.CSS_SELECTOR

- 사용 : 드라이버객체.find_element(By.XPATH, '//*[@id="account"]/div/a')

#### 찾은 요소에 대한 작업 실행
- find_element() 메소드로 찾은 요소에 대해 특정 명령을 내릴 수 있다.

- click() 해당 요소 클릭 : 드라이버객체.find_element(By.XPATH, '//*[@id="account"]/div/a').click()
- text 해당 요소에 있는 값 가져오기 : 드라이버객체.find_element(By.XPATH, '//*[@id="account"]/div/a').text
- send_keys() 해당 요소에 값 입력 : 드라이버객체.find_element(By.XPATH, '//*[@id="account"]/div/a').send_keys('문자열')

```python
from selenium import webdriver

from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options

from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

from webdriver_manager.chrome import ChromeDriverManager

import time

#################################################################################

customService = Service(ChromeDriverManager().install())
customOption = Options()

browser = webdriver.Chrome(service = customService, options = customOption)


URL = 'https://www.naver.com/'
browser.get(URL)
browser.implicitly_wait(10)

#################################################################################


#메일 값 획득
temp = browser.find_element(By.XPATH, '//*[@id="shortcutArea"]/ul/li[1]/a/span[2]').text
print(temp)


#만약 조금 상위 레벨에서 진행
temp = browser.find_element(By.XPATH, '//*[@id="shortcutArea"]/ul/li[1]/a').text
print(temp)


#send_keys
browser.find_element(By.XPATH, '//*[@id="query"]').send_keys('aaaaaa')
time.sleep(5)


# #버튼 클릭
# browser.find_element(By.XPATH, '//*[@id="account"]/div/a').click()
# time.sleep(3)
```

```python
import sys
import pymysql

def mysqlDbConnection(u, pw, h, p, d):
    try:
        conn = pymysql.connect(user = u, password = pw, host = h, port = p, database = d, charset='utf8')
        print(f"DB Connection Success: {h}")
    except pymysql.Error as e:
        print(f"Error connecting to MySQL Platform : {e}")
        sys.exit(1)
    return conn


def mysqlDbClose(_dbConn):
    try:
        _dbConn.close()
        print("DB Close Success")
    except pymysql.Error as e:
        print("Error closing from MySQL Platform")
        sys.exit(1)
```

```python
import random

def generate_random_decimal(s,e):
    return round(random.uniform(s, e), 1)

# 함수를 호출하여 무작위 소수점 첫째 자리 숫자 생성
random_decimal = generate_random_decimal(1,5)
print(random_decimal)
```

# 네이버 쇼핑인사이트 TOP 500 크롤링
### 크롤링 목표

네이버 데이터랩 : 쇼핑인사이트 - 분야 통계 - 인기검색어 TOP 500 크롤링

분야 선택 후 → 조회하기 → 분야에 대한 TOP500 순위 : 검색어 값 크롤링

크롤링한 값은 엑셀로 저장

#드롭 다운 박스 목록 보기 버튼 클릭

드라이버객체.find_element(By.XPATH, '//*[@id="content"]/div[2]/div/div[1]/div/div/div[1]/div/div[1]/span').click()

#변수에 값과 동일한 text를 가지는 요소 클릭

드라이버객체.find_element(By.XPATH, f'//*[@id="content"]/div[2]/div/div[1]/div/div/div[1]/div/div[1]/ul/li/a[contains(text(), "{변수}")]').click()

TOP500의 경우 한페이지에 20개의 검색어만 등장한다. 따라서 20개의 데이터를 긁고 버튼 클릭을 통해 다음 페이지로 넘어가야 한다.

```python
from selenium import webdriver

from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options

from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

from webdriver_manager.chrome import ChromeDriverManager

import time

import openpyxl

#################################################################################

customService = Service(ChromeDriverManager().install())
customOption = Options()

browser = webdriver.Chrome(service = customService, options = customOption)


URL = 'https://datalab.naver.com/shoppingInsight/sCategory.naver'
browser.get(URL)
browser.implicitly_wait(10)

#################################################################################

class1 = '식품'
class2 = ''
class3 = ''
class4 = ''

#class1
browser.find_element(By.XPATH, '//*[@id="content"]/div[2]/div/div[1]/div/div/div[1]/div/div[1]/span').click()
browser.find_element(By.XPATH, f'//div[2]/div/div[1]/div/div/div[1]/div/div[1]/ul/li/a[contains(text(), "{class1}")]').click()

#class2
if class2 != '':
    browser.find_element(By.XPATH, '//*[@id="content"]/div[2]/div/div[1]/div/div/div[1]/div/div[2]/span').click()
    browser.find_element(By.XPATH, f'//div[2]/div/div[1]/div/div/div[1]/div/div[2]/ul/li/a[contains(text(), "{class2}")]').click()

#class3
if class3 != '':
    browser.find_element(By.XPATH, '//*[@id="content"]/div[2]/div/div[1]/div/div/div[1]/div/div[3]/span').click()
    browser.find_element(By.XPATH, f'//div[2]/div/div[1]/div/div/div[1]/div/div[3]/ul/li/a[contains(text(), "{class3}")]').click()

#class4
if class4 != '':
    browser.find_element(By.XPATH, '//*[@id="content"]/div[2]/div/div[1]/div/div/div[1]/div/div[4]/span').click()
    browser.find_element(By.XPATH, f'//div[2]/div/div[1]/div/div/div[1]/div/div[4]/ul/li/a[contains(text(), "{class4}")]').click()


#조회하기 클릭
browser.find_element(By.XPATH, '//*[@id="content"]/div[2]/div/div[1]/div/a').click()
time.sleep(3)

#################################################################################

#엑셀 모듈에서 파일 -> 시트 객체 생성
xlsxFile = openpyxl.Workbook()
xlsxSheet = xlsxFile.active

# 20*25 = 500
#TOP500 긁기
for i in range(25):
    for j in range(1, 21, 1):
        keyword = browser.find_element(By.XPATH, f'//*[@id="content"]/div[2]/div/div[2]/div[2]/div/div/div[1]/ul/li[{j}]/a').text
        print(keyword)
        keyword = keyword.split('\n')

        xlsxSheet.cell(row = (i*20 + j), column = 1).value = (i*20 + j)
        xlsxSheet.cell(row = (i*20 + j), column = 2).value = keyword[1]

    browser.find_element(By.XPATH, '//*[@id="content"]/div[2]/div/div[2]/div[2]/div/div/div[2]/div/a[2]').click()
    time.sleep(3)


#엑셀 파일 저장
xlsxFile.save('naver_shopping_result.xlsx')
```
